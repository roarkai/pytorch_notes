# Double backward with custom functions
**使用场景**：计算高阶梯度

## I. 支持double backward的Custom Function
- **实现double backward的条件：**<font color=green>double backward本质上就是对backward中的computation再做一次backward。把backward看做一个函数，只要它的inputs中有tenosr是requires_grad=True的，那么autograd就能track 该函数中的computation，因而可以自动求梯度。</font>

- **backward的inputs分析**：有两个构成，一个是来自前序backward函数的grad_out，它的requires_grad默认是False。另一个是saved_tensor。

- **3种saved_tensor及其对应的处理方式：**
  1. **saved_tensor中有原函数的inputs，且其中有tensor的requires_grad=True**
     - saved_tensor向backward method传入了requires_grad=True的tensor input。所以autograd可以自动计算高阶梯度，自定义函数的时候不需要特殊处理。
  2. **saved_tensor中有原函数的outputs，且其requires_grad=True**
     - saved_tensor向backward method传入了requires_grad=True的tensor input。所以autograd可以自动计算高阶梯度，自定义函数的时候不需要特殊处理。
  3. <font color=blue>**saved_tensor中只有intermediate results，这些tensor的requires_grad在forward的计算中没有被自动设置为True**</font>
     - <font color=green>这里saved_tensor没有向backward method传入合适的input，为了能让autograd machine自动工作，需要一些额外的调整：</font>
       1. 手动将backward要用的intermediate results加入forward的output，此时他们作为forward的output会自动加上requires_grad=True
       2. 由于forward中多了本身不需要的output，为了遵守autograd的rule：“backward的input number与forward的output number必须一致”，在backward的inputs中对应加上与intermediate数量相同的参数。
       3. 由于此时forward的输出在形式上有多个（假设有k个），这意味着，forward function的inputs有k条path到final scalar output，因此在计算梯度的时候，要对这k条path的梯度做累加。

### I.1 saved_tensor是inputs
1. 如果<font color=green>**input tensor的requires_grad=True**</font>，且input tensor是saved_tensor，也就是backward method的input。在autograd的规则下，backward method的input tensor中至少有1个的requires_grad=True，因此backward method的output.grad_fn会指向backward computation的backward method，也就是二阶backward method。
2. 通过上述方式，autograd可以自动完成二阶梯度的计算。此时，torch.autograd.grad实际上被视为普通的function，它执行的computation对应的backward method也被记录到graph上。
3. 如果要用torchviz显示graph，要在torch.autograd.grad中设置<font color=green>**create_graph=True**</font>。否则运行grad之后，graph就会被释放掉。

```python
## 例1：机制示例  
class Square(torch.autograd.Function):
    @staticmethod
    def forward(ctx, x):
        ctx.save_for_backward(x)
        return x**2

    @staticmethod
    def backward(ctx, grad_out):
        x, = ctx.saved_tensors
        return grad_out * 2 * x

# finite differencing method会放大error，所以用双精度
x = torch.rand(3, 3, requires_grad=True, dtype=torch.double)

print(torch.autograd.gradcheck(Square.apply, x))

# 用gradcheck来验证second-order derivatives
print(torch.autograd.gradgradcheck(Square.apply, x))
```

#### I.1.1 用torch.autograd.grad画图
- 此时graph上有backward computation对应的backward method(double backward)的Node。如：下例中Mulbackward0。
  - square运算的backward method中唯一的computation是multiple。 
  - 所以double backward就是backward of multiple function。

```python
x = torch.tensor(1., requires_grad=True).clone()
y = Square.apply(x)

# 这里要设create_graph=True，否则后面torchviz画图会报错：
# RuntimeError: Trying to backward through the graph a second time
# (or directly access saved tensors after they have already been freed).
grad_x, = torch.autograd.grad(y, x, create_graph=True) 

# 完整的graph
torchviz.make_dot((grad_x, x, y), {'grad_x':grad_x, 'x':x, 'y':y}, show_saved=True)
# 简化：不画x作为output的那部分
# torchviz.make_dot((grad_x, y), {'grad_x':grad_x, 'y':y}, show_saved=True)
```

#### I.1.2 用autograd.backward画图：<font color=red>不要用这种方式！</font>
1. 图上没有backward computation对应的Node：因为backward没有output，因此没有output.grad_fn指向backward method。所以autograd就不会为该运算在graph上生成对应的Node
2. 更大的问题是可能导致memory leak：用backward() with create_graph=True 会导致reference cycle between the parameter and its gradient，会导致memory leak.

#### I.1.3 扩展：backward中也有自定义的saved tensor

```python
class Cube(torch.autograd.Function):
    @staticmethod
    def forward(ctx, x):
        ctx.save_for_backward(x)
        return x**3
    @staticmethod
    def backward(ctx, grad_out):
        x, = ctx.saved_tensors
        # 下面不用Square，直接用x**2的话，saved_tensor不会在图中显示
        return grad_out * 3 * Square.apply(x) 

x3 = torch.tensor(1., requires_grad=True).clone()
out3 = Cube.apply(x3)
grad_x3, = torch.autograd.grad(out3, x3, create_graph=True)

# 完整图
torchviz.make_dot((grad_x3, x3, out3), {"grad_x3": grad_x3, "x3": x3, "out3": out3}, show_saved=True)
# torchviz.make_dot((grad_x3, out3), {"grad_x3": grad_x3, "x3": x3, "out3": out3})
```

### I.2 saved_tensor是outputs
处理机制和前面saved_tensor是inputs相似。

```python
class Exp(torch.autograd.Function):
    # Simple case where everything goes well
    @staticmethod
    def forward(ctx, x):
        # This time we save the output
        result = torch.exp(x)
        # Note that we should use `save_for_backward` here when
        # the tensor saved is an ouptut (or an input).
        ctx.save_for_backward(result)
        return result

    @staticmethod
    def backward(ctx, grad_out):
        result, = ctx.saved_tensors
        return result * grad_out

x = torch.tensor(1., requires_grad=True, dtype=torch.double).clone()
# Validate our gradients using gradcheck
torch.autograd.gradcheck(Exp.apply, x)
torch.autograd.gradgradcheck(Exp.apply, x)
```

```python
out = Exp.apply(x)
grad_x, = torch.autograd.grad(out, x, create_graph=True)
torchviz.make_dot((grad_x, x, out), {"grad_x": grad_x, "x": x, "out": out})
```

### I.3 saved_tensor是Intermediate results
1. <font color=red>**如果要保存forward computation的中间结果传给backward使用。要将intermediate tensor作为forward method的outputs。**</font>
   - 当intermediate tensor成了forward的output，他们会被autograd自动加上requires_grad=True。<font color=blue>相当于转化成了前面第二种情况，saved_tensor是forward的output。</font>
2. 为了遵守autograd的rule：“backward的input number与forward的output number必须一致”，在backward的inputs中对应加上与intermediate数量相同的参数。
3. 由于此时forward的输出在形式上有多个（假设有k个），这意味着，forward function的inputs有k条path到final scalar output，因此在计算梯度的时候，要对这k条path的梯度做累加。

例如：implement sinh(x)函数 \
$sinh(x) = \frac{e^x-e^{-x}}{2}$  \
$\frac{dsinh(x)}{dx} = cosh(x) = \frac{e^x+e^{-x}}{2}$ \
从公式上可见，应该为backward的计算保留forward中的$e^x$和$e^{-x}$

```python
class Sinh(torch.autograd.Function):
    @staticmethod
    def forward(ctx, x):
        expx = torch.exp(x)
        expnegx = torch.exp(-x)
        ctx.save_for_backward(expx, expnegx)
        # In order to be able to save the intermediate results, a trick is to
        # include them as our outputs, so that the backward graph is constructed
        return (expx - expnegx) / 2, expx, expnegx

    # 注：backward的input tensor数量要和forward中的output tensor相同
    # _grad_out_exp和_grad_out_negexp分别对应的是expx和expnegx从上游拿到的grad_out
    # 在直接求backward时他们的值都是0，在求double backward时他们有具体值
    @staticmethod
    def backward(ctx, grad_out, _grad_out_exp, _grad_out_negexp):
        expx, expnegx = ctx.saved_tensors
        grad_input = grad_out * (expx + expnegx) / 2
        # 不会影响直接求导，因为此时_grad_out_exp和_grad_out_negexp都是0
        # 但需要这个‘形式’来确保double backward计算顺利
        grad_input += _grad_out_exp * expx
        grad_input -= _grad_out_negexp * expnegx
        return grad_input

# 要额外构建一个wrapper，让函数只返回第一个output
def sinh(x):
    return Sinh.apply(x)[0]

x = torch.rand(3, 3, requires_grad=True, dtype=torch.double)
torch.autograd.gradcheck(sinh, x)
torch.autograd.gradgradcheck(sinh, x)
```

```python
out = sinh(x)
grad_x, = torch.autograd.grad(out.sum(), x, create_graph=True)

torchviz.make_dot((grad_x, x, out), params={"grad_x": grad_x, "x": x, "out": out})
```

## II. 理解double backward的工作方式

1. 将自定义Function的backward method视为和forward一样的普通运算。backward的input arguments中的dout参数都视为requires_grad=False的外部参数。
2. 如果backward的参数就是forward的output，而不是ctx。那他们完全可以视为由两组计算构成的一整个计算。second backward就直接求这个‘整体’的backward就行。
3. 这里的问题是，从forward向backward传递的信息是通过ctx间接传递所需信息。好在虽然他们不直接是forward的output，也全部来自forward。所以，做一个简单处理，将ctx中的intermediate value全部设置为forward的output。那就可以将forward和backward两组计算重新视为一个整体了。
