# Autograd
## I. Autograd
### I.1 autograd是什么
Autograd是一个自动求微分的系统。
### I.2 autograd计算梯度的流程
1. **forward pass**\
（1）run operation to compute resulting tensor\
（2）记录operation的gradient function（形成抽象的DAG data structure）
2. **backward pass**: 调用<root_tensor>.backward()时执行\
（1）用graph上的.grad_fn计算gradients\
（2）将每个path上的梯度计算结果accumulate到对应tensor的.grad attribute\
（3）使用chain rule，向leaf tensor传梯度

<font color=red>**对accumulate的说明：** </font> 每个input tensor(eg: input=x)可能通过不同的path影响output。计算梯度的时候，dout/dx = dpath1/dx + dpath2/dx + ... + dpathn/dx。accumulate是指加总这些不同path的梯度。

## II. DAG: directed acyclic graph
1. DAG中的Node对象是torch.autograd.graph.Node：\
（1）所有requires_grad=True的input tensor(leaf)会有对应的Node。 \
（2）执行forward pass时，如果function的inputs中至少有1个tensor的requires_grad=True，那么该function的output的grad_fn会指向其backward method，此时backward method是以Node object的形式被refer to。即：
<font color=green>isinstance(output.grad_fn, torch.autograd.graph.Node)返回True</font>。
2. DAG的叶节点是input tensor，根节点是output tensor，在执行backward pass时，通过从root到leaf的方向tracing此图,就可以用链式法则自动计算梯度。

```python
import torch
import torchviz

a = torch.tensor([2., 3.], requires_grad=True)
b = torch.tensor([6., 4.], requires_grad=True)
Q = 3 * a ** 3 - b ** 2

# 这里Q不是scalar，所以backward要将output的dim作为参数
external_grad = torch.tensor([1., 1.])
Q.backward(gradient=external_grad)

# 画图
torchviz.make_dot(Q, {"a":a, "b":b})
```

## III. Autograd实现梯度计算的详细过程
### III.1 Forward pass中，autograd如何构建DAG并完成运算
1. **执行函数运算** \
（1）tensor调用的function是**torch.autograd.Function**的实例。这些function class中都定义了forward和backward函数。\
（2）这些function object都有apply()method[从torch.autograd.Function继承而来]。forward pass中，autograd会invoke function.apply(input)来执行函数运算。

2. **构建DAG** \
· forward pass过程中，autograd system在invoke function时，还会用torch.autograd.graph.Node，及Node之间的refer关系来表达DAG。具体方式：\
    i. 如果function的inputs中有tensor的requires_grad=True，那么autograd会为该tensor创建一个Node。Node会存储一个reference to tensor和其他信息。\
    ii. 同样，此时function的output作为intermediate tensor会被autograd自动设置为requires_grad=True。<font color=orange>**他们在DAG上不生成nodes。**</font>\
    iii. 在function的forward computation结束时（不用等到所有submodule的forward function全部执行完），autograd让<font color=blue>**function的每个output tensor中requires_grad=True的tensor的.grad_fn属性**</font>指向function的backward method。 \
    iv. 上述让grad_fn指向backward_method的动作，相当于在Graph中添加了Node。因为，ouput_tensor.grad_fn指向的对象是torch.autograd.graph.Node的子类。在没有添加Node时，output.grad_fn=None。<font color=red>[见下例]</font>。当root tensor上invoke了backward() method之后，autograd会根据链式法则，从后向前逐个调用DAG上tensor的grad_fn对应的backward method来计算梯度。

3. **DAG是动态图** \
每个迭代都会从零开始重新创建图。这种设计是为了让forward pass中的运算过程可以使用任意Python control flow statements。因为一旦有了control flow statements，每次迭代中图的形状和大小就可能发生变化。每次迭代都重新创建图的好处是，在启动训练之前,不必编码所有可能的path。实际运行了哪些函数，最后就求哪些函数对应的梯度。

```python
## 用forward hook on module打印forward执行时output的grad_fn

import torch.nn as nn
def forward_pre_hook(m, inputs):
    x, = inputs
    print('Before:', x.grad_fn)
    if x.grad_fn is not None:
        print('Name:', x.grad_fn.name, 
              isinstance(x.grad_fn, torch.autograd.graph.Node))
    return inputs

def forward_hook(m, inputs, output):
    print('After:', output.grad_fn)
    print('Name:', output.grad_fn.name, 
          isinstance(output.grad_fn, torch.autograd.graph.Node))
    print('-' * 50)
    return output

class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.aff1 = nn.Linear(3, 3)
        self.relu = nn.ReLU()
        self.aff2 = nn.Linear(3, 1)
    def forward(self, x):
        x = self.aff1(x)
        x = self.relu(x)
        x = self.aff2(x)
        return x
    
torch.manual_seed(2)
x = torch.randn(5, 3)
net = Net()

net.aff1.register_forward_pre_hook(forward_pre_hook)  
# input of aff1 没有ref to backward， 所以grad_fn=None
net.aff1.register_forward_hook(forward_hook)
# output则ref to backward，所以grad_fn=AddmmBackward0

net.aff2.register_forward_pre_hook(forward_pre_hook)
# aff2的input是relu的output，所以grad_fn=ReluBackward0

net.aff2.register_forward_hook(forward_hook)
# output则ref to AddmmBackward0

_ = net(x)
```

```python
x = torch.tensor([1., 2.])
target = torch.tensor([5., 7.])
w = torch.tensor([3.], requires_grad=True)
bias = torch.tensor([5.], requires_grad=True)
```

### III.2 如何从FP向BP传递信息
1. pytorch自带的函数，已经设置好了backward所需信息的保存方式。
2. 自定义nn.autograd.Function时，在forward method中用save_for_backward()来存储backward中会用到的tensor data。这些数据在backward method中可以用saved_tensors() method提取。pytorch中已经定义好的函数也是通过这种方式向backward method传递data的。
3. 需要的时候，比如debug，可以查看grad_fn对应的saved tensors。这些tensor以grad_fn的attribute形式存在，attribute name以'_saved'开头。
4. pytorch在saving时，会将function中的saved tensor打包，在读取他们的时候再unpack。但在读取他们的时候可能会因为要防止reference cycles，将tensor unpack成另一个tensor。此时读取出来的tensor和原来的tensor很可能不是同一个tensor。但是他们有共同的data storage。<font color=orange>2.1版本中，如果tensor是其grad_fn的output，就需要unpack为另一个tensor。但这个implementation dettails在未来可能会变化。</font>
5. 可以通过hooks for saved tensors来控制pytorch如何packing/unpacking saved tensors.<font color=red>[详见hooks专题]<>/font

```python
# 下图可以看到前面例子中autograd自动保存的forward ops中的saved information
torchviz.make_dot(loss, params={'w': w, 'bias': bias, 'x': x, 'diff': diff}, 
                  show_attrs=True, show_saved=True)
```

```python
# save input of function
a = torch.randn(5, requires_grad=True)
b = a.pow(2)
print(a.equal(b.grad_fn._saved_self)) # a不是它的grad_fn的output
print(a is b.grad_fn._saved_self)     # 所以a不用unpack成另一个tensor
```

```python
# save output of function
a = torch.randn(5, requires_grad=True)
b = a.exp()
print(b.equal(b.grad_fn._saved_result)) # b是它的grad_fn的output
print(b is b.grad_fn._saved_result)     # 所以b要unpack成另一个tensor
```

### III.3 在backward pass中计算梯度
1. loss.backward()会invokeDAG上backward对应的函数。forward pass中的Func，input，output与对应的backward函数关系：\
· <font color=lightblue>**output = Func(x)**</font> \
· <font color=lightblue>**dx = output.grad_fn(dout), output.grad_fn -> FuncBackward**</font>

```python
print('  loss.grad_fn ->', loss.grad_fn)
print('sqdiff.grad_fn ->', sqdiff.grad_fn)
print('  diff.grad_fn ->', diff.grad_fn)
print('     y.grad_fn ->', y.grad_fn)
print('scaled.grad_fn ->', scaled.grad_fn)
```

2. 这些backward method的invoke方式是：\
dsqdiff = SumBackward(sqdiff.shape).apply(1.) \
ddiff = PowBackward(base: diff, power: 2).apply(dsqdiff) \
dy = SubBackward().apply(ddiff) \
dscaled, dbias = AddBackward().apply(dy) \
dw = MulBackward(other: X).apply(dscaled)
3. 只有计算图中leaf nodes(requires_grad设为True时)才有grad properties，其他nodes都没有。
4. 一个计算图默认只能做一次BP。这是pytorch基于性能考虑设置的规则。如果要多次BP，要在backward call中设置参数retain_graph=True。

```python
## invoke backward之后，autograd会根据链式法则计算梯度
loss.backward()
print(w.grad)
print(bias.grad)
print(x.grad)
```

## IV. locally disabling gradient tracking/computation
默认所有requires_grad=True的tensor都会被track computational history。但有的时候，不需要计算梯度，为了提高运算效率，可以disable gradient computation。

· **不需要做gradient computation的典型场景：** \
  (1) model已经train好了，只做inference，不需要好用资源track history\
  (2) frozen参数来做finetune

· **diable gradient tracking的两类方法：** \
  (1) disable entire block of code用context manager,比如：no-grad mode和inference mode\
  (2) 在更精细的范围内exclusion of sub-gragphs from gradient computation可以通过设置tensor的requires_grad值。

### 方法1：setting requires_grad
1. tensor的requires_grad可以用来控制forward和backward中的gradient computation是否发生。requires_grad的默认是false，但wrapped in nn.Parameter的tensor的默认值是True。\
(1)Forward pass中，只有当operation中有tensor的requires_grad=True的时候才会被record到backward graph上。\
(2)Backward pass: 只有requires_grad=True的leaf tensor才会计算gradients并累积到tensor.grad上。所以，尽管每个tensor都设定了requires_grad的值，实际上只有leaf tensor的这个设定是有意义的。
2. 因为这个设置很常用，所以pytorch还支持在module上设置该参数，方法：nn.Mudule.requires\_grad_().但是这个设置是全局的，会对module中所有parameters发生作用。
3. leaf tensor和non-leaf tensor的区别：\
(1)leaf: 没有grad_fn; 没有backward graph与之对应；requires_grad的值要设定，要么手动设定，要么nn.Parameter中自动设定。\
(2)non-leaf:有grad_fn; 有backward graph与之对应，因此他们的gradient只是作为intermediate result来计算requires_grad=True的leaf tensor的gradient；也因此，所有non-leaf tensor都自动设定了requires_grad=True，不需要手动设定。\
**· <font color=orange>适用场景：</font>** training中需要update的那些leaf tensor才要设置为True；如果参数是放在nn.Parameter中，那么默认已经设置，不需要另外设置；此外就是finetune model的时候，用来freeze the update of tensor。\
**· <font color=lightgreen>例子1, finetune时，freeze weights</font>:** 可以对他们的tensor设置.apply.requires\_grad\_(False)


### 方法2：setting grad mode
· pytorch中有3种grad model，他们决定了autograd处理gradient的方式：默认的是grad mode，此外还有no-grad mode和inference mode。他们都可以通过context manager和decorator来设置。
1. **<font color=red>grad mode</font>:** 如果没有设置为另外两个mode，那么默认就是grad mode。只有在该mode下，requires_grad的配置才能起作用，autograd才会record operations on the tensors。在另外两种mode下，requires_grad都会被overriden to be False。

2. **<font color=red>no-grad mode</font>:** 让autograd暂时不记录operations（不将operations记录到backward graph上），退出该mode回到grad mode后再根据requires_grad的值是True还是False决定是否记录。\
  **· <font color=orange>适用场景：</font>** 只是暂时中断tensor operation的autograd，但tensor在当前operation中的运算结果还会被用于grad mode下的operation。\
  **· <font color=lightgreen>例子1, optimizer: </font>** 此时，weights' update不需要被autograd记录下来，但更新后的weights的值要用到下一次迭代中。而参与下一次迭代的forward运算时，这些weights是在grad mode下作计算的。\
  **· <font color=lightgreen>例子2, torch.nn.init</font>** 初始化operation是不需要计算梯度的。但是初始化完成后的training过程中，这些weights执行forward运算都是在grad mode中。

3. **<font color=red>inference mode</font>:** 不是暂停让autograd记录tensor参与的operation，而是完全关闭autograd对operation的tracking。优点是tensor的运算速度会比no-grad mode更快。代价是inference mode下创建的tensor即使退出inference mode后，autograd也无法tracking它们参与的operations。\
  **· <font color=orange>适用场景：</font>** tensor参与的运算不需要记录到backward graph，并且这些运算中创建的tensor后续也不会用于其他需要被autograd记录的运算中去。\
  **· <font color=lightgreen>例子1, data processing: </font>** 不是模型的一部分，不做backward。\
  **· <font color=lightgreen>例子2, model evaluation：</font>** inference不做backward

4. **<font color=red>evaluation mode</font>:** 这个不是disable autograd的机制，但容易和上述机制混淆。\
从功能上看，module.eval()和前面的no-grad和inference mode完全是不同的东西。理论上，model.eval()的作用取决于model中使用的modules，以及他们是否定义了trainning specific behavior。\
比如：model中用了torch.nn.Dropout或者torch.nn.Batchnorm，他们的特点是在trainning和validation/test时执行的operation不一样，这种情况下，就要手动在trainning和val/test阶段分别调用model.eval(), model.train()。
tensor的detach method。\
· <font color=red>建议：不论有没有定义trainning specific behavior，都在trainning和val/test阶段分别调用model.eval(), model.train()。因为，model中使用的module即使在现在没有trainning specific behavior，未来可能会发生改变。</font>

## V. 自定义torch.autograd.Function

步骤：\
· step1: subclass torch.autograd.Function, implement forward(), setup_context(), and backward() methods. \
· step2: 在以ctx（context object）作为argument的method中设置好其内容。\
· step3: declare是否支持double backwards\
· step4: 做gradient check

### step1：subclass Function
1. forward
2. ctx可以在forward中通过将ctx作为argument来对其设置。如果forward中没有设置ctx，就要用setup_context()。setup_context()只负责设置ctx，不做其他运算。
3. backward定义了gradient formula。\
· backward() method的tensor arguments是upstream gradients，function的每个output对应1个upstream gradient所以backward函数的tensor arguments数量与本函数的output数量相同。\
· backward() method的输出是function inputs的gradient。所以，backward的output tensor数量就是function inputs的数量。如果有的input tensor不用求gradient，或者有的input不是tensor，也不用求gradient，那么它们对应的输出用python：None。如果forward()中有optional arguments，那么输出的数量可能超过function input的数量，此时，backward()输出的数量也会超过function inputs，但这些forward arguments对应在backward中的output都要取None。

### step2：use the function in ctx properly,确保Funtction在autograd engine中正确工作
1. ctx.save_for_backward():存储backward pass中要用的tensor。\
· 如果要存储的data不是tensor，只能直接用ctx\
· 如果要存储的tensor既不是function的input也不是output，那么function就可能不支持double backward
2. ctx.mark_dirty()：用来mark在backward pass中会发生in-place modify的function input。
3. ctx.mark_none_differentiable()：用来标记function outputs是否需要求梯度。默认所有differentiable output tensors都设置为requires gradient。而不可导的那些则设置为不会被设置为requiures gradient。
4. ctx.set_materialize_grads()：当output tensor的计算不依赖于input值时，可以通过not materialize grad tensors given to backward function来优化梯度的计算方式。默认设置是Ture。如果设置为False，python中的None不会在call backward之前被转化为全0的tensor，此时，要手动处理这些None objects。

### step3：如果自定义的Function不支持double backward，要用decorating with once_differentiable()来声明
### step4：用torch.autograd.gradcheck()验证梯度计算的正确性。方法是用backward function计算Jacobian matrix，将结果与使用数值计算的结果做比较

```python
## 例1：自定义LinearFunction
import torch
from torch.autograd import Function
class LinearFunction(Function):
  @staticmethod
  def forward(input, weight, bias):
    output = input @ weight.T
    if bias is not None:
      output += bias.unsqueeze(0).expand_as(output)
    return output

  @staticmethod
  # inputs是所有传给forward()的inputs构成的tuple
  # output是forward()的output
  def setup_context(ctx, inputs, output):
    input, weight, bias = inputs
    ctx.save_for_backward(input, weight, bias)

  @staticmethod
  # grad_output是backward()唯一的输出
  def backward(ctx, grad_output):
    # unpack所有的saved_tensors
    input, weight, bias = ctx.saved_tensors
    # 初始化所有inputs的gradient为None，因为additional trailing Nones are ignored
    # 因此，即使在Function有additional input的时候，return statement都可以很简单
    grad_input = grad_weight = grad_bias = None

    # 增加的条件验证只是为了避免不必要的运算
    if ctx.needs_input_grad[0]:
      grad_input = grad_output.mm(weight)
    if ctx.needs_input_grad[1]:
      grad_weight = grad_output.t().mm(input)
    if bias is not None and ctx.needs_input_grad[2]:
      grad_bias = grad_output.sum(0)

    return grad_input, grad_weight, grad_bias
```

```python
## 应用方式
#  1. 最好wrapped into function形式
def linear(input, weight, bias=None):
    return LinearFunction.apply(input, weight, bias)

#  2. 直接用
linear = LinearFunction.apply
```

```python
## 例2：自定义一个有non-tensor argument的函数
class MulContent(Function):
  @staticmethod
  def forward(tensor, constant):
    return tensor * constant

  @staticmethod
  # inputs是所有传给forward()的inputs构成的tuple
  # output是forward()的output
  def setup_context(ctx, inputs, output):
    tensor, constant = inputs
    ctx.constant = constant  # 以ctx.constant的形式存非tensor

  @staticmethod
  # grad_output是backward()唯一的输出
  def backward(ctx, grad_output):
    return grad_output * ctx.constant, None # constant的梯度返回值用None
```

```python
## 例3（优化上例）：calling set_materialize_grads(False):
class MulContent(Function):
  @staticmethod
  def forward(tensor, constant):
    return tensor * constant

  @staticmethod
  # inputs是所有传给forward()的inputs构成的tuple
  # output是forward()的output
  def setup_context(ctx, inputs, output):
    tensor, constant = inputs
    ctx.set_materialize_grads(False)
    ctx.constant = constant  # 以ctx.constant的形式存非tensor

  @staticmethod
  # grad_output是backward()唯一的输出
  def backward(ctx, grad_output):
    # must handle None grad_output tensor
    # 这样就可以skip不必要的计算，直接返回None
    if grad_output is None:
        return None, None

    return grad_output * ctx.constant, None
```

如果需要save在forward中计算的intermediate tenosr，他们要returned as output才能支持autograd system完成double backward。<font color=red>详见[note of double backward]</font>

```python
## 例4：define gradient formula for intermediate values
class MyCube(Function):
  @staticmethod
  def forward(x):
    # 需要save x for backward，所以要将其return as output
    dx = 3 * x ** 2
    result = x ** 3
    return result, dx

  @staticmethod
  # inputs是所有传给forward()的inputs构成的tuple
  # output是forward()的output
  def setup_context(ctx, inputs, output):
    x, = inputs          # 因为inputs是tuple，所以只有1个元素的时候也要用'x,'
    result, dx = output
    ctx.save_for_backward(x, dx)

  @staticmethod
  # grad_output是backward()唯一的输出
  def backward(ctx, grad_output, grad_dx):
    # 打印从上游拿到的gradient of output of forward function's
    print(grad_output, grad_dx)
    x, dx = ctx.saved_tensors
    result = grad_output * dx
    # 下面这一步仅在double backward中有效，因为backward中grad_dx=0
    result += grad_dx * 6 * x
    return result

# wrap MyCube in a function
def my_cube(x):
    result, _ = MyCube.apply(x)
    return result
```

```python
x = torch.tensor(2., requires_grad=True).clone()
y = my_cube(x)
print(y)
grad_x, = torch.autograd.grad(y, x, create_graph=True)
print(grad_x)
torchviz.make_dot((y), {'y':y, 'x':x})
```

## VI. 几种不同的计算梯度的方式

### VI.1 基本概念：Jacobian, vjp和jvp
1. <font color=blue>**Jacobian：**</font> \
input x, x.shape = n; \
output y, y.shape = m; \
jacobian的shape是(m, n)
2. <font color=blue>vjp是vector dot product Jacobian。用在backward propagation中。vector和output的形状相同。</font>理解：\
（1）v是external_grad，外部梯度。也就是说，v本身在实际使用环境中也是一个梯度。比如：取v=dloss/dy，就可以理解成vjp将dloss/dx的求解分解成：vector @ Jacobian = dloss/dy @ dy/dx = dloss/dx \
（2）实际上v可以是任意值，只要和output形状相同即可。\
可以一般化成：loss=(y * v).sum()  <font color=green># 前提是v与x无关。</font> \
此时，dloss/dy = v  \
因此，dloss/dx = dloss/dy @ dy/dx =  vector @ Jacobian \
（3）典型的v取值是v=torch.ones_like(y)，此时相当于对loss = y.sum()求梯度

3. <font color=blue>jvp用在forward propagation中。vector和input的形状相同。</font>

<font color=red> **注：这里的@就是dot product，不做broadcast**</font>

```python
# 例1 Jacobian：x和y都是vector

def mulat(x):
    w = torch.ones(3, 2)
    return x @ w

x1 = torch.rand(3)
j1 = torch.autograd.functional.jacobian(mulat, x1)
print(j1.shape)
print(j1)
```

```python
# 对应求vjp
v1 = torch.ones(2)
torch.autograd.functional.vjp(mulat, x1, v1)
```

```python
# 例2.1 Jacobian：x是matrix，y是vector

def sum_row(x):
    return x.sum(dim=1)

x2 = torch.randn(2, 3)
y2 = sum_row(x2)
j2 = torch.autograd.functional.jacobian(sum_row, x2)
print('y', y2.shape)
print('x', j2, j2.shape)
```

```python
# 对应求vjp
v2 = torch.ones(3)
torch.autograd.functional.vjp(exp_reducer, x2, v2)
```

## VI.2 torch.autograd.backward和torch.autograd.grad


**几个关系：**
1. torch.autograd.backward和torch.autograd.grad都直接调用C++ engine to run the backward pass
2. tensor.backward是用torch.autograd.backward实现的
3. vjp，jvp都是用torch.autograd.grad来实现的

### torch.autograd.backward
**signature：** <font color=green>torch.autograd.backward(tensors, grad_tensors=None, retain_graph=None, create_graph=False, inputs=None)</font>
1. 计算tensors(tensor sequence)相对graph上叶节点的梯度之和。这里求和的含义是，如果tensors是vector而不是scalar，就将他们各自对叶节点的梯度求出后加总。
2. 如果tensors是vector而不是scalar，则还要指定grad_tensors为shape与tensors相同的vector。<font color=orange>[详见前文vjp和Jacobian的说明]</font>
3. retain_graph默认为create_graph的取值。如果设为True则backward完成后不释放graph。一般都用不到，也尽量不要用，通常都用空间效率更高的方式
4. create_graph也尽量不要用，容易leak memory。如果要用，就用torch.autograd.grad来实现该功能。
5. <font color=red>backward没有返回值，它计算出来的梯度会直接累加到leaf.grad上，optimaizer可以直接使用。</font>

### torch.autograd.grad
**signature：** <font color=green>torch.autograd.grad(outputs, inputs, grad_outputs=None, retain_graph=None, create_graph=False, only_inputs=True, allow_unused=None, is_grads_batched=False, materialize_grads=False)</font>
1. 计算outputs(tensor sequence)相对inputs的梯度之和。这里求和的含义是，如果outputs是vector而不是scalar，就将他们各自对叶节点的梯度求出后加总。
2. 如果outputs是vector而不是scalar，则还要指定grad_outputs为shape与outputs相同的vector。也就是vjp中的external_grad vector。<font color=orange>[详见前文vjp和Jacobian的说明]</font>
3. retain_graph默认为create_graph的取值。如果设为True则backward完成后不释放graph。一般都用不到，也尽量不要用，通常都用空间效率更高的方式
4. create_graph设为True时，会给derivative构建对应的graph。默认是False，需要计算高阶导数的时候可以设为True。<font color=orange>[详见Double Backward]</font>
5. <font color=red>grad有返回值，直接返回梯度计算结果，不会累加到input.grad上。因此，也无法直接被optimizer使用。</font>

**torch.autograd.grad**

```python
## 当output是sequence of tensors
x = torch.randn(4, requires_grad=True)
w1 = torch.ones(4)
w2 = torch.ones(4) * 0.5
y1 = (x * w1).sum()
y2 = (x * w2).sum()
# 如果outputs是tensor sequence，就是计算他们的gradient之和
# 注意返回值是tuple
grad, = torch.autograd.grad((y1, y2), x)
print(grad)
```
